# Dataset Exctraction to Zip

```python
import os
import shutil
import zipfile
import pandas as pd
import requests

# Mount Google Drive to access files
from google.colab import drive
drive.mount('/content/drive')

# Define the folder path to save the PDFs
download_folder = '/content/downloaded_pdfs'
zip_file_path = '/content/drive/Shareddrives/Automates (IL4)/Models/Predictive Space Scoring/Dataset/Research Papers/Full PDFs 2.zip'

# Create the temporary download folder
if not os.path.exists(download_folder):
  os.makedirs(download_folder)

# Read the CSV file containing the links for the first 1000 rows
file_path = '/content/drive/Shareddrives/Automates (IL4)/Models/Predictive Space Scoring/Dataset/Research Papers/data.csv'
df = pd.read_csv(file_path, nrows=1000)

# Function to download PDFs from a given URL
def download_pdf(pdf_link, download_folder):
  file_name = pdf_link.split('/')[-1]
  pdf_response = requests.get(pdf_link)
  if pdf_response.status_code == 200:
      with open(os.path.join(download_folder, file_name), 'wb') as f:
        f.write(pdf_response.content)
      print(f"PDF '{file_name}' downloaded successfully")
  else:
      print(f"Failed to download PDFs from {file_name}")

# Loop through each link in the dataframe and download PDFs
for link in df['link']:
  pdf_links = [link for link in link.split() if link.endswith('.pdf')]
  for pdf_link in pdf_links:
    download_pdf(pdf_link, download_folder)

# Create a zip file of the downloaded PDFs
with zipfile.ZipFile(zip_file_path, 'w') as zipf:
  for file_name in os.listdir(download_folder):
    file_path = os.path.join(download_folder, file_name)
    zipf.write(file_path, os.path.basename(file_path))

# Remove temporary folder
shutil.rmtree(download_folder)

print(f"PDFs zipped and saved to '{zip_file_path}'")
```

```python
import os
import pandas as pd
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_file):
    # Open the PDF file
    with fitz.open(pdf_file) as doc:
        text = ""
        # Iterate through each page in the PDF
        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            # Extract text from the page
            text += page.get_text()
    return text

def extraction(folder_path):
    # Get list of files in the folder
    files = os.listdir(folder_path)
    data = []
    # Iterate through each file
    for file_name in files:
        if file_name.endswith('.pdf'):
            file_path = os.path.join(folder_path, file_name)
            # Extract text from the PDF
            text = extract_text_from_pdf(file_path)
            # Append file name and extracted text to the data list
            data.append([file_name, text])
    # Create DataFrame from the data
    df = pd.DataFrame(data, columns=['File Name', 'Text'])
    return df

# Example usage:
folder_path = '/content/drive/Shareddrives/Automates (IL4)/Models/Predictive Space Scoring/Dataset/Research Papers/Extracted PDFs'
df = extraction(folder_path)
print(df)
```

```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
from transformers import BertTokenizer, TFBertModel

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = TFBertModel.from_pretrained("bert-base-uncased")

def vectorize_text(text):
  # Tokenize the text using BERT tokenizer
  tokenized_text = tokenizer.encode_plus(text, add_special_tokens=True, max_length=128, padding='max_length', truncation=True, return_tensors='tf')

  # Generate BERT embeddings
  outputs = bert_model(tokenized_text)
  embeddings = tf.reduce_mean(outputs.last_hidden_state, axis=1)

  return embeddings.numpy().tolist()
```

```python
embeddings = []
for i in df_arvix.index:
  embeddings.append(vectorize_text(str(df_arvix.loc[i, "Text"])))
```

```python
import pandas as pd
df_arvix = pd.read_csv("/content/drive/Shareddrives/Automates (IL4)/Models/Predictive Space Scoring/Dataset/scored_subset.csv")
```

```python
from ast import literal_eval
embeddings = []
for i in df_arvix.index:
  embeddings.append(literal_eval(df_arvix.loc[i, "embedding"]))
df_arvix["embedding"] = embeddings
```

```python
from math import sqrt
def euclidean_distance(vector1, vector2):
    if len(vector1) != len(vector2):
        raise ValueError("Vectors must have the same length")

    sum_squared_diff = sum((x - y) ** 2 for x, y in zip(vector1, vector2))
    return sqrt(sum_squared_diff)
```

```python
expected_scores = []
for j in df_arvix.index:
  # Initialize a list to store distances and indices
  distances = []

  # Loop through indices of df_arvix
  for i in df_arvix.index:
      if i != j:
          distance = euclidean_distance(df_arvix.loc[j, "embedding"][0], df_arvix.loc[i, "embedding"][0])
          distances.append((i, distance, df_arvix.loc[i, "score"]))  # Store index and distance as tuple

  # Sort distances list based on distance (using the second element of each tuple)
  distances.sort(key=lambda x: x[1])

  # Select the smallest 5 distances and their corresponding indices
  smallest_distances = distances[:5]

  # Print the smallest distances and their corresponding indices
  expected_scores.append(sum([smallest_distances[i][2] for i in range(5)])/5)
```

```python
df_arvix["expected_scores"] = expected_scores
```
